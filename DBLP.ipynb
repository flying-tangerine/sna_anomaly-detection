```python
import networkx as nx
import json
import cada
```

```python
with open(r'D:\Leiden\SNA\tmp_dblp_coauthorship.json', 'r') as f:
    data = json.load(f)
```

```python
G = nx.Graph()
edgeList = []
for item in data:
    timestamp=item[2]
    timeDic = {'timestamp': timestamp}
    edgeList.append((item[0], item[1], timeDic))
G.add_edges_from(edgeList)
```

```python
DiG = nx.DiGraph()
DiG.add_edges_from(edgeList)
```

```python
MultiG = nx.MultiGraph()
MultiG.add_edges_from(edgeList)
```

```python
print(G.number_of_nodes())
print(G.number_of_edges())
```

```python
print(DiG.number_of_nodes())
print(DiG.number_of_edges())
```

```python
DBLP_cadaL_t5 = cada.cada(G).get_anomalies_threshold(5)
```

```python
intG = nx.convert_node_labels_to_integers(G, label_attribute='origin')
```

```python
DBLP_cadaI_t5 = cada.cada(intG, algorithm='infomap').get_anomalies_threshold(5)
```

```python
len(DBLP_cadaI_t5)
```

```python
DBLP_cadaI_t5_restored = []
intNameDic = nx.get_node_attributes(intG, 'origin')
for i in DBLP_cadaI_t5:
    DBLP_cadaI_t5_restored.append(intNameDic[i])
```
