```python
import networkx as nx
import json
import cada
from collections import Counter
import math
```

```python
def Jaccard_similarity(c1, c2):

    n_union = len(set(c1).union(c2))
    n_intersection = len(set(c1).intersection(c2))
    
    return n_intersection / n_union
```

# Amazon

```python
with open(r'D:\Leiden\SNA\Amazon0601.txt', 'r') as f:
    edgeList = []
    for line in f:
        if line.startswith("#"):
            continue
        else:
            l = line.strip().split('\t')
            edgeList.append(tuple(l))
G_amazon = nx.Graph(edgeList)
DiG_amazon = nx.DiGraph(edgeList)
```

```python
n_amazon = G_amazon.number_of_nodes()
m_amazon = G_amazon.number_of_edges()
n_amazon_di = DiG_amazon.number_of_nodes()
m_amazon_di = DiG_amazon.number_of_edges()
print(n_amazon, m_amazon)
print(n_amazon_di, m_amazon_di)
```

## OddBall

```python
# int_G = nx.convert_node_labels_to_integers(G, label_attribute='origin')
nx.set_edge_attributes(G_amazon, values = 1, name = 'weight')
nx.write_edgelist(G_amazon, r'..\oddball_py3-master\network0_amazon.txt')
with open(r'..\oddball_py3-master\network_amazon.txt', 'w') as f:
    with open(r'..\oddball_py3-master\network0_amazon.txt', 'r') as fp:
        for line in fp:
            head, sep, tail = line.partition(' {')
            f.write(head)
            f.write(" " + tail[-3])
            f.write('\n')
f.close()
fp.close()
```

```python
with open(r'..\oddball_py3-master\code\output_amazon', 'r') as fd:
    count = 0
    anomalies_odd_amazon = []
    for line in fd:
        l = line.strip().split(" ")
        anomalies_odd_amazon.append(l[0])
        count += 1
        if count == int(n_amazon/100):
            break
fd.close()
```

## CADA

```python
Amazon_cadaL = cada.cada(G_amazon).get_top_anomalies(int(n_amazon/100))
```

```python
int_G_amazon = nx.convert_node_labels_to_integers(G_amazon, label_attribute='origin')
Amazon_cadaI = cada.cada(int_G_amazon, algorithm='infomap').get_top_anomalies(int(n_amazon/100))
Amazon_cadaI_restored = []
intNameDic_amazon = nx.get_node_attributes(int_G_amazon, 'origin')
for i in Amazon_cadaI:
    Amazon_cadaI_restored.append(intNameDic_amazon[i])
```

```python
counter_amazon_cadaL = Counter(Amazon_cadaL)
counter_amazon_cadaI = Counter(Amazon_cadaI_restored)
counter_amazon_odd = Counter(anomalies_odd_amazon)

print(Jaccard_similarity(counter_amazon_cadaL, counter_amazon_cadaI))
print(Jaccard_similarity(counter_amazon_cadaL, counter_amazon_odd))
print(Jaccard_similarity(counter_amazon_cadaI, counter_amazon_odd))
print(len(set(counter_amazon_cadaL).intersection(counter_amazon_cadaI).intersection(counter_amazon_odd)))
print(set(counter_amazon_cadaL).intersection(counter_amazon_cadaI).intersection(counter_amazon_odd))
```

# DBLP

```python
with open(r'D:\Leiden\SNA\tmp_dblp_coauthorship.json', 'r') as f:
    data_dblp = json.load(f)
f.close()
```

```python
print(data_dblp[0], data_dblp[-1])
```

```python
G_dblp = nx.Graph()
edgeList = []
for item in data_dblp:
    timestamp=item[2]
    timeDic = {'timestamp': timestamp}
    edgeList.append((item[0], item[1], timeDic))
G_dblp.add_edges_from(edgeList)
```

```python
DiG_dblp = nx.DiGraph()
DiG_dblp.add_edges_from(edgeList)
```

```python
n_dblp = G_dblp.number_of_nodes()
m_dblp = G_dblp.number_of_edges()
n_dblp_di = DiG_dblp.number_of_nodes()
m_dblp_di = DiG_dblp.number_of_edges()
print(n_dblp, m_dblp)
print(n_dblp_di, m_dblp_di)
```

## OddBall

```python
int_G_dblp = nx.convert_node_labels_to_integers(G_dblp, label_attribute='origin')
nx.set_edge_attributes(int_G_dblp, values = 1, name = 'weight')
nx.write_edgelist(int_G_dblp, r'..\oddball_py3-master\network0_dblp.txt')
with open(r'..\oddball_py3-master\network_dblp.txt', 'w') as f:
    with open(r'..\oddball_py3-master\network0_dblp.txt', 'r') as fp:
        for line in fp:
            head, sep, tail = line.partition(' {')
            f.write(head)
            f.write(" " + tail[-3])
            f.write('\n')
f.close()
fp.close()
```

```python
with open(r'..\oddball_py3-master\code\output_dblp', 'r') as fd:
    count = 0
    anomalies_odd_dblp = []
    for line in fd:
        l = line.strip().split(" ")
        anomalies_odd_dblp.append(l[0])
        count += 1
        if count == int(n_dblp/100):
            break
fd.close()
```

```python
anomalies_odd_dblp_restored = []
intNameDic_dblp = nx.get_node_attributes(int_G_dblp, 'origin')
for int_node in anomalies_odd_dblp:
    anomalies_odd_dblp_restored.append(intNameDic_dblp[int(int_node)])
print(len(anomalies_odd_dblp_restored), anomalies_odd_dblp_restored)
```

## CADA

```python
DBLP_cadaL = cada.cada(G_dblp).get_top_anomalies(int(n_dblp/100))
DBLP_cadaL
```

```python
DBLP_cadaI = cada.cada(int_G_dblp, algorithm='infomap').get_top_anomalies(int(n_dblp/100))
```

```python
DBLP_cadaI_restored = []
for i in DBLP_cadaI:
    DBLP_cadaI_restored.append(intNameDic_dblp[i])
DBLP_cadaI_restored
```

```python
counter_cadaL = Counter(DBLP_cadaL)
counter_cadaI = Counter(DBLP_cadaI_restored)
counter_odd = Counter(anomalies_odd_dblp_restored)

print(Jaccard_similarity(counter_cadaL, counter_cadaI))
print(Jaccard_similarity(counter_cadaL, counter_odd))
print(Jaccard_similarity(counter_odd, counter_cadaI))
print(len(set(counter_cadaL).intersection(counter_cadaI).intersection(counter_odd)))
print(set(counter_cadaL).intersection(counter_cadaI).intersection(counter_odd))
```
